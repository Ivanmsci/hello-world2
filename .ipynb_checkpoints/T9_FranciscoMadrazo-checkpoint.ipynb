### TAREA 9

Simular el sistema masa-resorte ahora considerando un amortiguador.


Los valores de los parámetros son $m=0.5\, kg$, $k = 0.8$ y $b = 0.4$. Condiciones iniciales $x(0)=0.2\,m$, $v(0)=0\,m/s$ y en un intervalo de tiempo de $t=[0,30]\, s$

El sistema de ecuaciones diferenciales que modela el sistema es:

\begin{align}
\frac{dx}{dt} &= v\\
\frac{dv}{dt} &= -\frac{k}{m}x - \frac{b}{m}v
\end{align}

        

```python
# importar librerías
import numpy as mp
import matplotlib.pyplot as plt
import scipy.integrate as ode
```

```python
#Definir ecuación diferencial
def masa_resorte(t,y):
    x = y[0]
    v = y[1]
    
    #Parámetros
    m = 0.5
    k = 0.8
    b = 0.4
    #Definir el sistema de ecuaciones diferenciales
    dx = v
    dv = (-(k/m)*x) - ((b/m)*v)
    
    return [dx,dv]
```

```python
#Condición inicial
y0 = [0.2,0]

#Intervalo de tiempo
t1 = (0,30)

#solución
sol_masa_resorte = ode.solve_ivp(masa_resorte,t1,y0,max_step=0.1)
sol_masa_resorte
```

```python
#Obtener vector t (tiempo) y solución (y)
t = sol_masa_resorte.t
y = sol_masa_resorte.y.T
y
```

```python
#Graficas
plt.figure(figsize=(15,10))
plt.plot(t,y,lw=3,label=['Posicion del sistema masa-resolrte','Velocidad del sistema masa-resorlte'])
plt.xlabel('$Tiempo [s]$', fontsize=15)

plt.legend(loc='best')
plt.grid()
plt.show()
```
