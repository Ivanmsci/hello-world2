```python
import numpy as np

# importar matplotlib.pyplot

import matplotlib.pyplot as plt

# comando para que las gr√°ficas salgan en la misma ventana}

%matplotlib inline
```

```python
#silicio

Eg01 = 1.17
beta1 = 636
alfa1 = 4.73*(10**-4)
def f_si(T):
    return Eg01 - (alfa1*(T**2))/(T + beta1)

#GaAs

Eg02 = 1.521
beta2 = 220
alfa2 = 5.58*(10**-4)
def f_GaAs(T):
    return Eg02 - (alfa2*(T**2))/(T + beta2)

#Ge

Eg03 = 0.74
beta3 = 235
alfa3 = 4.774*(10**-4)
def f_Ge(T):
    return Eg03 - (alfa3*(T**2))/(T + beta3)
```

```python
Tnum = np.linspace(0,800,80)


ynum1 = f_si(Tnum)
ynum2 = f_GaAs(Tnum)
ynum3 = f_Ge(Tnum)

# Graficar


plt.figure(figsize=(15,8))

# Crear el tama√±o de la figura a graficar
plt.plot(Tnum, ynum1, '--b', lw = 3, label = '$Silicio$')
plt.plot(Tnum, ynum2, '--p', lw = 3, label = '$GaAs$')
plt.plot(Tnum, ynum3, '--k', lw = 3, label = '$Ge$')


#Nombrar los ejes de la gr√°fica
plt.xlabel ('$T(K)$')
plt.ylabel('$Eg = Eg(0) - (alfa*T)/(beta + T) $')


#mostrar la cuadricula

plt.legend(loc = 'best')    #acomoda las eqtiquetas donde al programa le convenga XD

plt.grid()     #muestra la cuadricula

plt.show()     #muestra lo que se ha hecho

```
